```python
import matplotlib.pyplot as plt
```

```python
no_function = open('mostra/sem funçao - mostra.csv')
```

```python
second = []
volt = []
linha = []
for i in no_function:
    linha = i.split(',')
    second.append(linha[0])
    volt.append(float(linha[1]))
```

```python
n1 = []

for i in range(len(second)):
    n1.append(i)

plt.figure(figsize = (20, 20))
plt.xticks(rotation = 'vertical')
plt.bar(n1[2:], volt[2:], color='red')
plt.title('Sem função alguma')
```
