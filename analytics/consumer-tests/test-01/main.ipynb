```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
del data['Unnamed: 2']

df = data.iloc[1:]
df
```

```python
consum = [float(x) * 1000 for x in df['X(CH1)'].values]
```

```python
time = [float(x) + 30.399998 for x in df['Time'].values]
```

```python
serie = pd.Series(consum, index=time)
serie
```

```python
plt.figure(figsize=(16, 8))

plt.ylabel('Energy consumption (mA)')
plt.xlabel('Time in secondos')

# 274 samples per minute
init=round(274 * 1 + 6)
final=round(274 * 3)
plt.plot(time[init:final], consum[init:final], '-.')
```
